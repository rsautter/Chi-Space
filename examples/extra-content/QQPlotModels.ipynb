```python
import scipy
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
scipy.__version__
```

```python
gauss = stats.norm()
```

```python
sga = gauss.rvs(4096)
sga =sga - np.average(sga)
sga = sga/np.std(sga)
plt.figure(figsize=(20,5))
plt.plot(sga)
plt.xlim(0,4096)
plt.show()
```

```python
gumb = stats.gumbel_r()
```

```python
sg = gumb.rvs(4096)
sg =sg - np.average(sg)
sg = sg/np.std(sg)
plt.figure(figsize=(20,5))
plt.plot(sg)
plt.xlim(0,4096)
plt.show()
```

```python
wei = stats.weibull_min(0.9)
```

```python
sw = wei.rvs(4096)
sw = sw - np.average(sw)
sw = sw /np.std(sw)
plt.figure(figsize=(20,5))
plt.plot(sw)
plt.xlim(0,4096)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,3,1)
plt.hist(sga,bins=20,color='blue')
plt.title("Gaussian Distribution")

ax = plt.subplot(2,3,4)
stats.probplot(sga,plot = ax)

plt.subplot(2,3,2)
plt.hist(sga,bins=20,color='blue')
plt.title("Gumbel Distribution")
ax = plt.subplot(2,3,5)
stats.probplot(sg,plot = ax)

plt.subplot(2,3,3)
plt.hist(sw,bins=20,color='blue')
plt.title("Weibull Minimum Distribution (c=0.9)")
ax = plt.subplot(2,3,6)
stats.probplot(sw,plot = ax)
plt.tight_layout()
plt.show()
```
